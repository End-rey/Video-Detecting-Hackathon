```python
import os
import torchvision.transforms as tt
import torchvision.io
from torchvision.utils import save_image
path_to_photo = 'weapon-detection-6/train/images/'
path_to_label = 'weapon-detection-6/train/labels/'
dr = os.listdir(path_to_photo[:-1])
for i in range(len(dr)):
    with open(path_to_label + dr[i][:-4] + '.txt', 'r') as f:
        j = 0
        line = f.readline()
        while len(line) > 0:
            if line[0] == '0':
                tlbr = [float(x) for x in line.split()[1:]]
                image = torchvision.io.read_image(path_to_photo + dr[i])
                width = image.size()[1]
                height = image.size()[2]
                image = tt.functional.resized_crop(image, top=round((tlbr[1]-tlbr[3]/2)*height), left=round((tlbr[0]-tlbr[2]/2)*width),
                                                  width=round(tlbr[2]*width), height=round(tlbr[2]*height), size=width)
                save_image(image / 255, f'datasets/pose_ds/images/{i}_{j}.jpg')
                results = model.predict(f'datasets/pose_ds/images/{i}_{j}.jpg')[0].keypoints.xyn[0]
                if len(results) > 0:
                    zero_arr = torch.zeros(height, width)
                    for point in results:
                        zero_arr[min(torch.round(point[1] * height).cpu().numpy(), 639)][min(torch.round(point[0] * width).cpu().numpy(), 639)] = 1
                    save_image(zero_arr, f'datasets/pose_ds/points/{i}_{j}.jpg')
                    j += 1
            line = f.readline()
```
