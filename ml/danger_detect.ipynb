```python
import json
import os
import cv2

box_dict = {'gun_bb': 0, 'per_bb': 1}

with open('video_gun_gunner_labels.json') as json_file:
    data = json.load(json_file)
    for video in data:
        video_title = video['data_title']
        frames = video['data_units'][video['data_hash']]['labels']
        vidcap = cv2.VideoCapture(f'datasets/video/{video_title}')
        success, image = vidcap.read()
        count = 0
        while success:
            if frames.get(str(count + 1), False):
                if len(frames[str(count + 1)]['objects']) == 2:
                    image = cv2.resize(image, (640, 640), interpolation = cv2.INTER_AREA)
                    cv2.imwrite(f"datasets/video/images/{video_title[:-4]}{count}.jpg", image)
                    with open(f'datasets/video/labels/{video_title[:-4]}{count}.txt', 'w') as f:
                        s = ''
                        for x in frames[str(count + 1)]['objects']:
                            s += f'''{box_dict[x["name"]]} {x["boundingBox"]["x"] + x["boundingBox"]["w"]} {x["boundingBox"]["y"] + x["boundingBox"]["h"]} {x["boundingBox"]["w"]} {x["boundingBox"]["h"]}
                            '''
                        f.write(s)
            success,image = vidcap.read()
            count += 1
```
