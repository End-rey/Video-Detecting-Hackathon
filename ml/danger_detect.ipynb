```python
from ultralytics import YOLO
import torch
```

```python
model = YOLO('weights/yolov8n.pt')
```

```python
for k, v in model.named_parameters():
    print(k)
```

```python
freeze = [f'model.{x}.' for x in range(10)]  # layers to freeze
for k, v in model.named_parameters():
    v.requires_grad = True  # train all layers
    if any(x in k for x in freeze):
        print(f'freezing {k}')
        v.requires_grad = False
```

```python
import json
import os
 
# Opening JSON file
with open('labels.json') as json_file:
    data = json.load(json_file)
    for x in data[1:]:
        if len(x['data_units'][x['data_hash']]['labels']['objects']) != 0:
            try:
                os.rename(f'../firearms detection.v2i.yolov8/train/images/{x["data_title"]}', f'train/images/{x["data_title"]}')
            except FileNotFoundError:
                os.rename(f'../gun.v4i.yolov8/train/images/{x["data_title"]}', f'train/images/{x["data_title"]}')
            finally:
                with open(x['data_title'][:-4] + '.txt', 'w') as f:
                    for y in range(len(x['data_units'][x['data_hash']]['labels']['objects'])):
                        temp = x['data_units'][x['data_hash']]['labels']['objects'][y]['boundingBox']
                        f.write(f'{y} {temp["x"]} {temp["y"]} {temp["w"]} {temp["h"]}')
```

```python
import json
with open('labels.json') as json_file:
    data = json.load(json_file)
    for x in data:
        if len(x['data_units'][x['data_hash']]['labels']['objects']) != 0:
            with open('datasets/train/labels/' + x['data_title'][:-4] + '.txt', 'w') as f:
                for y in range(len(x['data_units'][x['data_hash']]['labels']['objects'])):
                    temp = x['data_units'][x['data_hash']]['labels']['objects'][y]['boundingBox']
                    f.write(f'0 {temp["x"] + temp["w"] / 2} {temp["y"] + temp["h"] / 2} {temp["w"]} {temp["h"]}')
```
